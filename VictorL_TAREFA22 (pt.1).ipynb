#### 1. Baixando dados de ativos
    1. Baixe os dados de três ativos da sua escolha do Yahoo Finance, correspondente a um período de pelo menos 2 anos até a data mais recente.
    2. Salve esses dados para que você possa utilizá-los na próxima tarefa.

```python
########################## BLOCO DE IMPORTAÇÕES ##########################
import pandas as pd
import numpy as np
import plotly.express as px
import yfinance as yf

########################## BLOCO DE INICIALIZAÇÕES #######################
ativos = ['ZS=F', 'ZL=F', 'ZMH24.CBT']
dt_ini = '2022-01-01'
columns = ['data', 'ativo', 'valor']
amp = pd.DataFrame(columns=columns)

########################## BLOCO DE ABERTURA #############################
yf_data = yf.download(ativos, start=dt_ini)

########################## BLOCO DE TRANSFORMAÇÃO ################
ativo_mapping = {
    'ZL=F': 'Oleo',
    'ZMH24.CBT': 'Farrinha',
    'ZS=F': 'Grao'
}

yf_data = yf_data.rename(columns=ativo_mapping, level=1)

########################## BLOCO DE EXIBIÇÃO ############################
yf_data
```

#### 2. Observando a volatilidade dentro do dia
1. Calcule a amplitude de variação do valor do ativo dentro do dia. Essa amplitude você pode calcular com um comando simples de adicionar uma coluna no Pandas: (valor máximo - valor mínimo).
2. Monte um gráfico de linha mostrando a variação dessa amplitude ao longo do período coletado, por ativo.
3. Veja se consegue identificar que ativo apresentou maior volatilidade.

```python
########################## BLOCO DE ENGENHARIA ###########################
high = yf_data['High']
high_vert = high.stack().reset_index()
high_vert.columns = columns

low = yf_data['Low']
low_vert = low.stack().reset_index()
low_vert.columns = columns

amp['data'] = high_vert['data']
amp['ativo'] = high_vert['ativo']
amp['valor'] = round(high_vert['valor'] - low_vert['valor'], 2)

########################## BLOCO DE EXIBIÇÃO ############################
amp
```

```python
########################## BLOCO DE EXIBIÇÃO #############################
fig = px.line(amp,
              x='data',
              y='valor',
              color='ativo',
              template='simple_white')
fig.show()
```

> Definitivamente, o **Grão** da Soja apresenta maior volatibilidade financeira ao longo do tempo.

De fato, o commodity bruto possui menor valor comercial e esta mais sujeito a variações comerciais. Já os subprodutos de processamento do Grão, agregam valor e estão subjeitos a um escopo modificado de variações; o que acaba sendo traduzido em maior estabilidade temporal do ativo.
