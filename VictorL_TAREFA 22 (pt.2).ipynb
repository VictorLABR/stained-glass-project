#### 1. Carregue os dados da tarefa anterior

````python
########################## BLOCO DE IMPORTAÇÕES ##########################
import pandas as pd
import numpy as np
import plotly.express as px
import yfinance as yf
import plotly.graph_objects as go

from plotly.graph_objects import Layout
from datetime import datetime, timedelta

########################## BLOCO DE INICIALIZAÇÕES #######################
ativos = ['ZS=F', 'ZL=F', 'ZMH24.CBT']
dt_ini = '2022-01-01'
columns = ['data', 'ativo', 'valor']
columns_agg = ['ativo', 'data', 'valor']

########################## BLOCO DE ABERTURA #############################
yf_data = yf.download(ativos, start=dt_ini)

########################## BLOCO DE TRANSFORMAÇÃO ########################
ativo_mapping = {
    'ZL=F': 'Oleo',
    'ZMH24.CBT': 'Farrinha',
    'ZS=F': 'Grao'
}

yf_data = yf_data.rename(columns=ativo_mapping, level=1)

########################## BLOCO DE EXIBIÇÃO ############################
yf_data
````

#### 2. Média móvel

1. Faça uma média móvel  em 90 dias dos valores de fechamento ajustados dos ativos que você escolheu
2. Utilizando a fução ```rolling()```, calcule o desvio pardão desses ativos ao longo do tempo, utilizando a mesma janela móvel do item acima
2. Monte um gráfico de linhas para as médias móveis dos ativos da sua base
4. Monte um gráfico de linhas para os desvios padrão em janela móvel
5. Compare os dois gráficos que você fez com o da tarefa anterior. Veja se eles apontam conclusões semelhantes quanto aos períodos de maior volatilidade (variação) dos ativos.

````python
########################## BLOCO DE ENGENHARIA ############################
fech = yf_data['Adj Close']
fech_vert = fech.stack().reset_index()
fech_vert.columns = columns

media_movel = fech_vert.groupby('ativo')['valor'].rolling(window=90).mean().reset_index()
media_movel.columns = columns_agg

desv_pad = fech_vert.groupby('ativo')['valor'].rolling(window=90).std().reset_index()
desv_pad.columns = columns_agg

########################## BLOCO GRÁFICO ############################
fig_media = px.line(media_movel, x='data', y='valor', color='ativo',
                    title='Médias Móveis de 90 dias',
                    labels={'Media Movel': 'Valor', 'data': 'Dias'},
                    template='simple_white')
fig_desv_pad = px.line(desv_pad, x='data', y='valor', color='ativo',
                       title='Desvios-Padrão em janela móvel de 90 dias',
                       labels={'Desvio-Padrao': 'Desvio-Padrão', 'data': 'Dias'},
                       template='simple_white')

########################## BLOCO DE EXIBIÇÃO ############################
fig_media.show()
````

````python
########################## BLOCO DE EXIBIÇÃO ############################
fig_desv_pad.show()
````

> A Média Móvel é muito mais estável temporalmente, logo, ela **não** apresenta muita variação ao longo das medições.

> No entanto, os Desvios-Padrão das Médias Móveis, apresentam **sim** um comportamento complementar (alternado) ao gráfico gerado na Tarefa anterior.

No caso, considerando o mesmo período, na mesma janela a amplitude tende a aumentar enquanto o desvio da média móvel tende a cair.

#### 3. Gráfico de *candlestick*
1. Selecione o período dos últimos 60 dias (corridos) para um dos ativos da sua base
2. Monte um gráfico de *candlestick* para esses dados.

````python
########################## BLOCO DE ENGENHARIA ###########################
data_atual = yf_data.index.max()
data_inicio = data_atual - timedelta(days=60)
df_grao = yf_data.xs('Grao', axis=1, level=1)

########################## BLOCO GRÁFICO ################################
grafico = {'x': df_grao.index,
           'open': df_grao.Open,
           'close': df_grao.Close,
           'high': df_grao.High,
           'low': df_grao.Low,
           'type': 'candlestick',
           'name': 'Grao',
           'showlegend': True
           }

fig = go.Figure(data=grafico,
                layout_title='Preços Futuros do Grão de Soja')

########################## BLOCO DE EXIBIÇÃO ############################
fig.show()
````
