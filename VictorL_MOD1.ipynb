```python
# ticket_médio é igual a valor_total / quantidade_total, logo, o primeiro cálculo:
valor_total_19= "(A)"
quantidade_total_19= 3
ticket_medio_19= 320.52

print('O valor em reais do total de vendas no dia 19/01 é:')
valor_total_19= ticket_medio_19 * quantidade_total_19
print(valor_total_19)
```

```python
# O segundo cálculo:
valor_total_20= 834.47
quantidade_total_20= "(B)"
ticket_medio_20= 119.21

print('A quantidade total de vendas no dia 20/01 é:')
quantidade_total_20= valor_total_20 / ticket_medio_20
print(quantidade_total_20)
```

```python
# Já o terceiro cálculo:
valor_total_21= 15378.12
quantidade_total_21= 5
ticket_medio_21= "(C)"

print('O ticket médio no dia 21/01 é:')
ticket_medio_21 = valor_total_21 / quantidade_total_21
print(ticket_medio_21)
```

```python
# definindo a variável do exercício
cancao='Roda mundo, roda gigante, roda moinho, roda pião.'

# primeiro método: caps lock
print('Normalizando a canção:')
cancao= cancao.upper()
print(cancao)

# segundo método: procura
print('A canção cita moinhos? *Valor -1 para negativo e valor n para n-ésima casa da canção;')
tem_moinho_na_cancao= cancao.find('MOINHO')
print(tem_moinho_na_cancao)

# terceiro método: substituição
print('Criando segunda estrofe da canção:')
segunda_estrofe= cancao.replace('RODA','GIRA')
print(segunda_estrofe)
```

```python
noticia= 'Selic vai a 2,75% e supera expectativas; ' + \
 'é a primeira alta em 6 anos.'

# como o enunciado pede o valor sem exigir qual classe específica:
print('Qual a taxa selic do exercício?')
posicao_valor_selic= noticia.find('2')
selic=noticia[posicao_valor_selic:posicao_valor_selic+5]
print(selic)

# idem:
print('É a primeira alta em quantos anos?')
posicao_ano= noticia.find('6')
ano=noticia[posicao_ano]
print(ano)
```
