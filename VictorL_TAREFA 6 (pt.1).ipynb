#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esses arquivos.

```python
# atualmente, este é meu bloco fixo de importações
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import requests as rq
%matplotlib inline # com adição deste

base = pd.read_csv("SINASC_RO_2019.csv") # carregando a Base

print(base.columns) # verificando o nome das variáveis da Base
print(base.shape) # verificando as dimensões da Base
```

####  2) Monte gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
fig, axes = plt.subplots(1, 5, figsize=(15, 3))
sns.set(style="whitegrid")

sns.countplot(ax= axes[0], x= "APGAR1", data= base)
sns.countplot(ax= axes[1], x= "APGAR5", data= base)
sns.countplot(ax= axes[2], x= "SEXO", data= base)
sns.countplot(ax= axes[3], x= "GESTACAO", data= base)
sns.countplot(ax= axes[4], x= "ESTCIVMAE", data= base)

fig.subplots_adjust(wspace=0.5)

for ax in axes:
    ax.tick_params(axis= "x", labelrotation= 90)

plt.show()
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
sns.barplot(x="APGAR5", y="CONSULTAS", data= base)

plt.xlabel("APGAR5 médio")
plt.ylabel("Número de consultas")
plt.title("APGAR5 médio por Consultas")

plt.show()
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

sns.countplot(ax= axes[0, 0], x= "SEXO", data= base)
sns.countplot(ax= axes[0, 1], x= "GESTACAO", data= base)
sns.countplot(ax= axes[1, 0], x= "ESTCIVMAE", data= base)
sns.countplot(ax= axes[1, 1], x= "CONSULTAS", data= base)

# Rotacionar os rótulos do eixo X do gráfico "GESTACAO" em 30 graus
axes[0, 1].set_xticklabels(axes[0, 1].get_xticklabels(), rotation=90)

fig.subplots_adjust(wspace=0.25, hspace= 0.9)

plt.show()
```
