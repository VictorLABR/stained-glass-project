> Lacunas: as variáveis **escolaridade**, **estado civil** e **salário anual** tem aproximadamente a média de 1127 dados faltando (cerca de 11% cada variável).


É necessário, antes de seguir com as próximas etapas, conferir o quão afetada a variável dependente será pela limpeza das lacunas.

```python
# construção da função que levanta estatísticas sobre as variáveis que possuem lacunas
def status_lacunas(df: pd.DataFrame) -> None:

  status_lacunas = []
  for col in df.columns:
    if df[col].isna().any():
      qtd, _ = df[df[col].isna()].shape
      total, _ = df.shape
      dict_lacunas = {col: {"quantidade": qtd, "porcentagem (%)": round(100 * qtd/total, 2)}}
      status_lacunas.append(dict_lacunas)

  for status in status_lacunas:
    print(status)
```

```python
status_lacunas(df=df)
```

```python
status_lacunas(df=df[df['default'] == 0])
```

```python
status_lacunas(df=df[df['default'] == 1])
```

> O subconjunto de dados com lacunas praticamente não interage com as proporções da variável dependente, adimplência.

> Sendo assim, é possível iniciar o tratamento de dados. Começando com a retirada das lacunas e corrigindo o levantamento dos tipos de entrada dos dados nas variáveis que estão incompatíveis com as saídas.

```python
df.shape # verificando a Base antes do tratamento
```

```python
df.dropna(inplace=True) # uso de uma funcionalidade do dataframe que retira as lacunas da Base de Dados
```

```python
# criação da função fn que altera a formatação únivoca dos dados que geram a incompatibilidade nos tipos de variáveis
fn = lambda valor: float(valor.replace(".", "").replace(",","."))

# limpeza das duas variáveis numéricas que apresentam entrada categórica de dados
df["valor_transacoes_12m"] = df["valor_transacoes_12m"].apply(fn)
df["limite_credito"] = df["limite_credito"].apply(fn)
```
