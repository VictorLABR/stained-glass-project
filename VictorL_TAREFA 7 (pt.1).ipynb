````python
# atualmente, este é meu bloco fixo de importações
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import requests as rq
%matplotlib inline

dados = pd.read_csv("demo01_mod7.csv")

# esse método apresenta simultaneamente todas variáveis, Não-Vazios, a contagem de Missings e os Tipos das variáveis
dados.info()

# como a coluna "mau" é booleana, inicialmente não é possivel gerar um histrograma sem tratamento prévio, logo:
tabela_freq = pd.crosstab(index=dados["mau"], columns = "count")
tabela_freq_percentual = tabela_freq.apply(lambda x: round(x / x.sum() * 100, 2), axis=0)
tabela_freq_percentual
# portanto, a distribuição da variável resposta "mau" é 97,66% falsa e 2,34% verdadeira
````

#### 2) Vamos montar um metadados

1. Crie um dataframe com os nomes de cada variável e o tipo de dados de cada variável.
2. Adicione uma coluna nesse *dataframe* chamada "qtd_categorias" e coloque nela o número de categorias correspondente de cada variável. 
    Dica: 
        1. inicie uma lista vazia
        2. faça um for ao longo dos nomes das variáveis, 
        3. conte o número de categorias dessa variável 
        4. acumule essa informação de 3. na lista que você criou em 1. 
        5. No final, essa lista pode ser atribuída à nossa variável.
3. Crie variáveis dummy para as variáveis necessárias (i.e. aquelas que são qualitativas e não estão armazenadas como {0, 1} ou {True, False}.

````python
# Criando um dataframe com os nomes e tipos de dados das variáveis
df_info = pd.DataFrame({"Variável": dados.columns, "Tipo": dados.dtypes})

numero_cat = []

# Iterando sobre as variáveis do novo dataframe
for variable in dados.columns:
    # Contando o número de categorias da variável - apenas para variáveis categóricas -
    if dados[variable].dtype == "object":
        numero_cat.append(dados[variable].nunique()) # incluindo apenas 1 tipo de categoria, sem repetição, caso exista...
    else:
        numero_cat.append(None)

# Adicionando a lista como uma nova coluna do dataframe de informações
df_info["qtd_categorias"] = numero_cat
print(df_info.to_string(index=False)) # Bibliotequei no pandas como não repetir as variáveis: usando de to_string sem o índice

# Criando variáveis dummy para as variáveis categóricas listadas pelo df_info
df_dummy = pd.get_dummies(dados, columns=["sexo",
                                          "posse_de_veiculo",
                                          "posse_de_imovel",
                                          "tipo_renda",
                                          "educacao",
                                          "estado_civil",
                                          "tipo_residencia"])

df_dummy.head()
````

#### 3) Crie variáveis dummy para as variáveis necessárias (i.e. aquelas que são qualitativas e não estão armazenadas como {0, 1} ou {True, False}. Crie um *dataframe* apenas com as variáveis apropriadas para entrada no scikitlearn - elimine as variáveis tipo *str*, mantendo apenas suas versões *dummy*.

````python
df_dummy.mau = df_dummy.mau.astype(int) # só havia restado a variável resposta para ser tratada
df_scikitlearn = df_dummy
df_scikitlearn.info()
````

#### 4) Qual variável é mais poderosa?

Considere as variáveis ```possui_email``` e ```posse_de_veiculo```. Faça uma tabela cruzada entre elas e responda qual delas te parece mais poderosa para prever a probabilidade de ```mau = 1```?

````python
# Criando tabela cruzada entre mau, possui_email e posse_de_veiculo_Y [porque posse_de_veiculo_N é apenas o complementar]
tabela_cruzada = pd.crosstab(index=df_scikitlearn["mau"],
                             columns=[df_scikitlearn["possui_email"], df_scikitlearn["posse_de_veiculo_Y"]],
                             margins=True,
                             normalize=True)

# Criando as proporções de mau = 1 entre as categorias de email e veiculo através da tabela cruzada
prop_email_sim = tabela_cruzada.loc[1, (1, 0)] + tabela_cruzada.loc[1, (1, 1)]
prop_email_nao = tabela_cruzada.loc[1, (0, 0)] + tabela_cruzada.loc[1, (0, 1)]
prop_veiculo_sim = tabela_cruzada.loc[1, (0, 1)] + tabela_cruzada.loc[1, (1, 1)]
prop_veiculo_nao = tabela_cruzada.loc[1, (0, 0)] + tabela_cruzada.loc[1, (1, 0)]

# Bloco de exibição dos resultados das comparações, convertidos para porcentagem com 2 casas após a vírgula
print(f"Proporção de mau pagador para possui email: {prop_email_sim:.2%}")
print(f"Proporção de mau pagador para não possui email: {prop_email_nao:.2%}")
print(f"Proporção de mau pagador para tem veiculo: {prop_veiculo_sim:.2%}")
print(f"Proporção de mau pagador para não tem veiculo: {prop_veiculo_nao:.2%}")
````

Portanto, apesar de baixas porcentagens, a proporção entre as proporções é de fato mais expressiva entre a variável EMAIL e a target MAU;

onde a posse de veículo fica próxima da metade, cerca de 0,58x, entre SIM & NÃO frente ao alvo,

mas a posse de email atinge exatamente 12x mais NÃOS do que SIMS frente a MAUS pagadores.
